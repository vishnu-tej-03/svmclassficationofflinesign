```python
#the identity matrix
np.identity(5)
```

```python
#repeat an array
arr=np.array([[1,2,3]])
r1=np.repeat(arr,3,axis=0)
print(r1)
```

```python
arr=np.array([[1,2,3]])
r1=np.repeat(arr,3,axis=1)
print(r1)
```

```python
output=np.ones((5,5))
print(output)

z=np.zeros((3,3))
z[1,1]=9
print(z)

output[1:4,1:4]=z
print(output)
```

##### Be careful while copying arrays!!!!

```python
a=np.array([1,2,3])
b=a
b[0]=100
print(b)
print(a)
##problem
```

```python
#solved
a=np.array([1,2,3])
b=a.copy()
b[0]=100
print(b)
print(a)
```

##### Reorganising Arrays

```python
before=np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after=before.reshape((8,1))
print(after)
```

```python
#vertically stacking vectors
v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])
np.vstack([v1,v2,v1,v2])
```

```python
#horizontal stack
np.hstack([v1,v2,v1])
```
