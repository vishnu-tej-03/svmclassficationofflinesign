##### Loading necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
#plot x,y 1-d arrays
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize graph (dpi specifies pixels per inch,use 300 for good quality)
plt.figure(figsize=(5,3),dpi=300)

#plt.plot(x,y,label='2x',color='green',linewidth=2,marker='.',linestyle='--' ,markersize=10,markeredgecolor='red')#name of plot

#shorthand
plt.plot(x,y,'b^--',label='2x')

#add line2
x2=np.arange(0,4.5,0.5)#select interval we want our point in
plt.plot(x2[:5],x2[:5]**2,'r.-',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'r.--')

#title of graph
plt.title('Our first graph!',
fontdict={'fontname':'Comic Sans MS','fontsize':20})

#x label and y label
plt.xlabel('X Axis',
    fontdict={'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('Y Axis',
    fontdict={'fontname':'Comic Sans MS','fontsize':15})

#only want integer values the graph will resize automatically
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#add a legend on the axis
plt.legend()

#save figure
plt.savefig('mygraph.png',dpi=200)

#display the graph
plt.show()
```

##### Bar chart

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)

patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))#from starting if nothing would have popped from the end

plt.figure(figsize=(6,4),dpi=300)

plt.show()
```
